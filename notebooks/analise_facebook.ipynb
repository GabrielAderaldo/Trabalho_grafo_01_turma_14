# Infraestrutura de Dados: Facebook Ego-Network

Este notebook fornece a infraestrutura de carga para os dados gerados pelo Java. 
As variáveis abaixo contêm os dados brutos prontos para análise.

---

```python
import pandas as pd
import numpy as np
import json
import os

# Definição de Caminhos
METRICS_JSON = '../data/generated/plain_text/facebook_union_metrics.json'
DEGREES_CSV = '../data/generated/sheets/facebook_union_vertex_degrees.csv'
EDGELIST_BIN = '../data/generated/bin/facebook_union_edgelist.bin'

print(">>> Infraestrutura carregada.")
```

## 1. Carga de Métricas Globais (Dicionário)
A variável `metrics` contém os valores calculados manualmente no Java.

```python
metrics = {}
if os.path.exists(METRICS_JSON):
    with open(METRICS_JSON, 'r') as f:
        metrics = json.load(f)
    print(f">>> [OK] Métricas globais carregadas: {list(metrics.keys())}")
else:
    print(">>> [ERRO] Execute 'make run-full' no Java primeiro.")
```

## 2. Carga de Dados Estruturais (NumPy)
A variável `edges` contém a lista de adjacência em formato de matriz NumPy (E x 2).

```python
edges = np.array([])
if os.path.exists(EDGELIST_BIN):
    data = np.fromfile(EDGELIST_BIN, dtype=np.int32)
    V_count, E_count = data[0], data[1]
    edges = data[2:].reshape((E_count, 2))
    print(f">>> [OK] Grafo carregado via NumPy: {V_count} vértices, {E_count} arestas.")
else:
    print(">>> [AVISO] Arquivo binário não encontrado.")
```

## 3. Carga de Estatísticas de Grau (Pandas)
A variável `df_degrees` contém os graus de cada um dos 4039 vértices.

```python
df_degrees = pd.DataFrame()
if os.path.exists(DEGREES_CSV):
    df_degrees = pd.read_csv(DEGREES_CSV)
    print(f">>> [OK] DataFrame de graus carregado: {df_degrees.shape[0]} linhas.")
else:
    print(">>> [AVISO] CSV de graus não encontrado.")
```
